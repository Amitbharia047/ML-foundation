```python
import numpy as np
```

```python
marks = np.array([[78,85,90],[45,60,55],[88,92,95],[30,40,35],[65,70,75]])
print(marks)
```

```python
print(np.shape(marks))  # Shape of the array
print(np.size(marks))   # Total number of elements
print(np.result_type(marks))  # Data type of the elements


print(marks[1]) # Accessing the second row (Student 2's marks)
print(marks[:, 2]) # marks of all students in the third subject
print(marks[0, 1])#  marks of first student in second subject
```
