manager decides to add 2 bonus unit to every product's daily sales for a holiday report Add 2 to the entire daily sales.
multiply the cost price array by the daily sales matrix.
observe how numpy handles the multiplication of 1d and 2d array.


```python
daily_sales_bonus = daily_sales + 2

print("\nAfter Adding Bonus Units:\n", daily_sales_bonus)

cost_price = np.array([2000, 50000, 8000])  # per unit

print("\nCost Price:\n", cost_price)

revenue_matrix = daily_sales_bonus * cost_price[:, np.newaxis]

print("\nRevenue Matrix:\n", revenue_matrix)

total_revenue = np.sum(revenue_matrix)

print("\nTotal Revenue =", total_revenue)
```

Where in numpy

```python
arr = np.array([10,-5,20,-3,15])
result = np.where(arr<0,0,arr)
print (result)
```

```python
arr = np.array([1,2,3,4,5,6])
# if element is odd we repplace with -1 else element itself
result= np.where(arr%2!= 0,-1,arr)
print (result)
```

```python
result= np.where(arr>=50,1,0)
print (result)
```

```python
data =np.array([10,20,30,40,50])
mean_val = np.mean(data)
result = np.where(data >mean_val,mean_val,data)
print(result)
```
