```python
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "Marks": [85, 90, 95]
}
print(data) #python dictionary
print(type(data))
#convert to pandas dataframe
marks = pd.DataFrame(data) #pandas automatically adds a default index starting from 0 by default. this is called RangeIndex

print(marks)
```

# Handling missing values

```python
df = pd.DataFrame({
    "Name": ["Aman","Sourav","Karan","shubham",None,"sameer"],
    "Age": [22,23,None,22,25,25],
    "Marks": [85,None,88,90,56,None]

})
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
cleaned_data=df.dropna()
```
