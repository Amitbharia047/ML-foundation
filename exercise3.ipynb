Task A1
Create a numpy array electricity using np.random.randint 
-- Values between 120 and 650
--shape(4,7)

```python
import numpy as np
```

```python
np.random.seed(42)   

electricity = np.random.randint(120, 651, size=(4, 7))

print(electricity)
```

Task B1: Hostel-wise Metrics
Find the avg electricity usage per hostel the maximum usage per hostel and identify which hostel has the highest average usage 

```python

avg = electricity.mean(axis=1)
mx = electricity.max(axis=1)
hostel = np.argmax(avg) + 1

print(avg)
print(mx)
print("Highest avg: H", hostel)
```

Task B2: Day-wise Metrics 
--Create the total electricity consuption per day and identify the day with the highest total compus consumption

```python
import numpy as np

total_per_day = electricity.sum(axis=0)
highest_day = np.argmax(total_per_day)

print(total_per_day)
print("Highest consumption day index:", highest_day)
```

--Values > 550 -> multiplied by 1.20

```python
penalized_usage = np.where(electricity > 550,
                           electricity * 1.20,
                           electricity)

print(penalized_usage)
```
